* 4번 : pop의 칼럼 인덱스를 구별, 인구수, 한국인, 외국인, 고령자 로 변경

```python
pop.columns=['구별', '인구수', '한국인', '외국인', '고령자']
```

```python
pop.head()
```

```python
pop.drop(0, inplace=True)
```

* 5번 : cctv데이터에서 '총계'로 내림차순 정렬

```python
cctv.sort_values('총계', ascending=False)
```

* 6번 : cctv 최근 증가율
    (2021 cctv증가수 + 2020 cctv증가수) / (2019 cctv) => 최근증가율 칼럼

```python
cctv['최근증가율'] = (cctv['2021년']+cctv['2020년'])/cctv['2019년']
```

```python
cctv
```

* 7번 : pop에 외국인비율, 고령자비율 추가 

    외국인비율 = 외국인/인구수100     
    고령자비율 = 고령자/인구수100

```python
pop['외국인비율'] = pop['외국인']/pop['인구수']*100
pop['고령자비율'] = pop['고령자']/pop['인구수']*100
```

```python
pop.head(3)
```

```python
cctv.head(3)
```

* 8번 : cctv의 구분을 구별로 변경

```python
cctv.rename(columns={'구분':'구별'}, inplace=True)
```

```python
cctv.head(2)
```

* 9번 : cctv와 pop을 구별을 기준으로 병합해서 data_result에 저장.

```python
data_result = pd.merge(cctv,pop,on='구별')
```

* 11번 : cctv, pop병합 단, cctv 총계, 최근증가율만 가져와서 병합

```python
data_result = pd.merge(cctv[['구별','총계','최근증가율']],pop,on='구별')
```

```python
data_result
```

* 12번 : data_result 인덱스를 구별로 설정

```python
data_result.set_index('구별', inplace=True)
```

## 상관관계 분석

-1 ~ 0 ~ 1(절대값)  
0.1 이하 : 관계없음. 무시  
0.3 이하 : 약한 상관 관계  
0.7 이하 : 뚜렷한 상관 관계  


### 고령자 비율, cctv 총계 상관관계

```python
np.corrcoef(data_result['고령자비율'], data_result['총계'])  #음수면 앞의거 적을수록 뒤에거가 크다는뜻
```

### 외국인비율, cctv 총계 상관관계

```python
np.corrcoef(data_result['외국인비율'], data_result['총계'])
```

### 인구수, cctv 총계 상관관계

```python
np.corrcoef(data_result['인구수'], data_result['총계'])
```
