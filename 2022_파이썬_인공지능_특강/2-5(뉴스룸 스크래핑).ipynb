# 이미지 다운로드

### 이미지 주소 확보

```python
div_img0 = soup.select('.news-img-box')[0]
```

```python
div_img0
```

```python
div_img0['style']
```

```python
div_img0['style'][div_img0['style'].find('(')+1:div_img0['style'].rfind(')')]
```

```python
imgTags = soup.select('.news-img-box')
imgUrl = []
for imgTag in imgTags:
    style = imgTag['style']
    imgurl = style[style.find('(')+1:style.rfind(')')]
    imgUrl.append(imgurl)
```

```python
imgUrl
```

# 이미지 다운로드

```python
img1 = requests.get(imgUrl[0], verify=False)
img1.raise_for_status()
```

```python
img1.content
```

```python
import matplotlib.pyplot as plt
```

```python
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
img123 = Image.open(BytesIO(img1.content))
plt.figure(figsize=(10,10),dpi = 200,facecolor='white')
plt.imshow(img123)
plt.axis('off')
plt.show()
```

```python
with open('1.jpg', 'wb') as f:   #이미지 저장
    f.write(img1.content)
```

```python
def downloadImg(filename, imgUrl):
    img = requests.get(imgUrl, verify=False)
    img.raise_for_status()
    with open(filename, 'wb') as f:
        f.write(img.content)
    print(f"{filename} 다운로드 완료 {len(img.content)/1000}kB")
```

```python
downloadImg('2.jpg', imgUrl[1])
```
