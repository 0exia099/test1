```python
#loss_DropOut.ipynb에 이어서
from tensorflow import keras
from sklearn.model_selection import train_test_split
```

```python
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

train_scaled = train_input /255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
def model_fn(a_layer=None):
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=(28,28)))
    model.add(keras.layers.Dense(100, activation='relu'))
    if a_layer:
        model.add(a_layer)
    model.add(keras.layers.Dense(10, activation='softmax'))
    return model
```

```python
model = model_fn(keras.layers.Dropout(0.3)) #load_weights()로 모델 복원시에는 구조가 동일해야 한다.
model.load_weights('model-weights.h5') #저장한 파라미터 복원
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) #load_weights()로 파라미터 복원 시 버그?로 인해 compile해야함(버그인지 불확실)
model.summary()
```

```python
model.evaluate(val_scaled, val_target)
```
