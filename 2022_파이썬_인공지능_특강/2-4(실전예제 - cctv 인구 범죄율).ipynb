```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Apple SD Gothic Neo'
plt.rcParams['font.size'] = 14
#plt.rcParams['axes.unicode_minus'] = False.  #한글폰트(New Gulim) -기호 해결
```

```python
data_result = pd.read_csv('data_result.csv', index_col = '구별')
```

```python
data_result.head()
```

* 1번 : cctv 총계를 내림차순으로 정렬해서 막대 그래프(barh)를 그려보시오

```python
data_result['총계'].sort_values(ascending=False).plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
data_result['총계'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

* 2번 : 인구수 대비 cctv 비율을 구해서 cctv비율 이라는 컬럼에 저장  
    cctv비율 = cctv수/인구수*100

```python
data_result['cctv비율'] = data_result['총계']/data_result['인구수']*100
```

```python
data_result
```

```python
data_result['cctv비율'].sort_values(ascending=False).plot(kind='bar', grid=True, figsize=(10,10))
plt.show()
```

* 3번 : Scatter 그래프를 사용해서, 인구수(x축), cctv총계(y축)의 관계를 그려보시오.

```python
plt.figure(figsize=(10,10),facecolor='white')
plt.scatter(data_result['인구수'], data_result['총계'], s=50)
plt.xlabel('인구수')
plt.ylabel('cctv 총계')
plt.grid()
plt.show()
```

* 4번 : 경향선을 하나 추가

```python
fp1 = np.polyfit(data_result['인구수'], data_result['총계'],1)
```

```python
fp1 #기울기, y절편
```

```python
f1 = np.poly1d(fp1)
```

```python
fx = np.linspace(100000, 700000, 2)
```

```python
plt.figure(figsize=(10,10),dpi = 200,facecolor='white')
plt.scatter(data_result['인구수'], data_result['총계'], s=50)
plt.plot(fx, f1(fx), 'r--') #경향선 추가
plt.xlabel('인구수')
plt.ylabel('cctv 총계')
plt.grid()
plt.show()
```

* 5번 : 오차 컬럼을 data_result에 추가  
    오차는 경향선과 실제 값의 차이

```python
data_result['오차'] = np.abs(data_result['총계']-f1(data_result['인구수']))
```

```python
data_result
```

```python
plt.figure(figsize=(10,10),dpi = 200,facecolor='white')
plt.scatter(data_result['인구수'], data_result['총계'], 
            s=data_result['외국인비율']*40, c=data_result['오차'])
plt.colorbar()
plt.plot(fx, f1(fx), 'r--') #경향선 추가
plt.xlabel('인구수')
plt.ylabel('cctv 총계')
plt.grid()
plt.show()
```

* 6번 : scatter 그래프에 각 점마다 구 이름을 추가하시오.(plt.text()사용)

```python
plt.figure(figsize=(12,12),dpi = 200,facecolor='white')
plt.scatter(data_result['인구수'], data_result['총계'], 
            s=data_result['외국인비율']*40, c=data_result['오차'])
plt.colorbar()
plt.plot(fx, f1(fx), 'r--') #경향선 추가

#구이름 추가
for guName in data_result.index:
    plt.text(data_result.at[guName, '인구수']*1.02, data_result.at[guName,'총계']*0.99,guName)
    
plt.xlabel('인구수')
plt.ylabel('cctv 총계')
plt.grid()
plt.show()
```

* 7번 : 구이름을 오차가 1000보다 큰 경우만 표시

```python
plt.figure(figsize=(12,12),dpi = 200,facecolor='white')
plt.scatter(data_result['인구수'], data_result['총계'], 
            s=data_result['외국인비율']*40, c=data_result['오차'])
plt.colorbar()
plt.plot(fx, f1(fx), 'r--') #경향선 추가

#구이름 추가
for guName in data_result.index:
    if data_result.at[guName,'오차'] >1000:
        plt.text(data_result.at[guName, '인구수']*1.02, data_result.at[guName,'총계']*0.99,guName)
    
plt.xlabel('인구수')
plt.ylabel('cctv 총계')
plt.grid()
plt.show()
```

## 범죄 현황도 함께 분석을 하자
서울시 5대 범죄 발생현황 통계

```python
crime = pd.read_csv('crime_in_seoul20220803.csv')
```

```python
crime.info()
```

```python
crime = pd.read_csv('crime_in_seoul20220803.csv', skiprows=3)
```

```python
crime.head(1)
```

```python
del crime['발생']
del crime['검거']
del crime['자치구별(1)']
crime.head(1)
```

```python
crime.columns = ['구별','살인발생','살인검거','강도발생','강도검거',
                 '강간강제추행발생','강간강제추행검거','절도발생','절도검거',
                 '폭력발생','폭력검거']
```

```python
crime.head(1)
```

```python
crime.drop(0,inplace=True)
```

```python
crime.set_index('구별', inplace=True)
```

```python
#각 범죄마다 검거율 추가
```

```python
crime.info()
```

```python
crime.replace('-','0',inplace=True)
```

```python
crime.info()
```

```python
crime['살인발생'] = crime['살인발생'].astype(np.int64)
crime['살인검거'] = crime['살인검거'].astype(np.int64)
```

```python
crime.info()
```

```python
for n in ['살인','강도','강간강제추행','절도','폭력']:
    crime[n+'검거율'] = crime[n+'검거']/crime[n+'발생']*100
```

```python
crime
```

* 검거율이 100보다 큰 경우 값을 100으로 조정

```python
for n in ['살인','강도','강간강제추행','절도','폭력']:
    crime.loc[crime[n+'검거율']>100,n+'검거율'] = 100
```

```python
crime
```

```python
#crime nan을 0으로
crime.fillna(0, inplace=True)
```

```python
crime
```

```python
crime.loc['도봉구','살인검거율'] = 100
```

```python
crime
```

```python
#각 검거 칼럼을 모두 삭제
for c in crime.columns:
    if c.endswith('검거'):
        del crime[c]
```

```python
crime
```

```python
# 칼럼이름에서 발생을 없애기 ex) 살인발생->살인
for c in crime.columns:
    if c.endswith('발생'):
        crime.rename(columns={c:c[:-2]},inplace=True)
```

```python
crime
```

```python
#데이터 정규화(normalize)
from sklearn import preprocessing
```

```python
x = crime[['살인', '강도', '강간강제추행','절도','폭력']]
```

```python
x
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
x_scaled = min_max_scaler.fit_transform(x.astype(float))
```

```python
x_scaled
```

```python
crime_norm = pd.DataFrame(x_scaled)
```

```python
crime_norm
```

```python
crime_norm = pd.DataFrame(x_scaled, index=crime.index, columns=crime.columns[:5])
```

```python
crime_norm
```

```python
#crime_norm에 crime의 각 검거율을 추가
for c in crime.columns:
    if '검거율' in c:
        crime_norm[c] = crime[c]
```

```python
crime_norm
```

```python
#cctv_result.csv로 cctv_result 데이터프레임으로(구별 행의 인덱스로)
cctv_result = pd.read_csv('cctv_result.csv', index_col = '구별')
```

```python
cctv_result.head(2)
```

```python
#crime_norm에 cctv_result의 인구수, 총계를 추가(단 총계를 cctv수 로 바꿔서)
crime_norm[['인구수', 'cctv수']] = cctv_result[['인구수', '총계']] 
```

```python
crime_norm
```

```python
#범죄 칼럼 만들기(범죄 = 모든 범죄의 발생 수의 합
crime_norm['범죄'] = crime_norm.loc[:,'살인':'폭력'].sum(axis=1)
```

```python
crime_norm
```

```python
#살인검거율 ~~ + ~~ 폭력검거율 다 더해서 검거에 저장
crime_norm['검거'] = crime_norm.loc[:,'살인검거율':'폭력검거율'].sum(axis=1)
```

```python
crime_norm
```

```python
import seaborn as sns
```

```python
sns.pairplot(crime_norm)
plt.show()
```

```python
sns.pairplot(crime_norm[['인구수','cctv수', '범죄', '검거']])
```

```python
sns.pairplot(crime_norm[['인구수','cctv수', '범죄', '검거']], kind='reg')
plt.show()
```

```python
sns.pairplot(crime_norm[['인구수','cctv수', '범죄', '검거']], kind='reg', plot_kws={'line_kws':{'color' : 'red'}})
plt.show()
```

```python

sns.pairplot(crime_norm,x_vars=['인구수','cctv수'], y_vars=['검거','살인','강도'], kind='reg')
plt.show()
```

```python
#검거의 최대값으로 각 검거의 값을 나누고 * 100을 해서 다시 검거에 저장
crime_norm['검거'] = crime_norm['검거']/crime_norm['검거'].max()*100
```

```python
crime_norm
```

```python
#crime_norm 검거의 내림차순으로 정렬한 후 crime_norm_sorted에 저장
crime_norm_sorted = crime_norm.sort_values('검거', ascending=False)
```
