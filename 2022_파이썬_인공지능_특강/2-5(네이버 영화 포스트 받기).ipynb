```python
def getTitlesandImgUrls(webUrl):
    r = requests.get(webUrl)
    if r.status_code != 200:
        return None
    soup = BeautifulSoup(r.content, 'html.parser')
    titles = [a.get_text() for a in soup.select('dt a')]
    imgUrls = [imgTag['src'].split('?')[0] for imgTag in soup.select('.result_thumb a img')]
    posterInfo = dict(zip(titles, imgUrls))
    return posterInfo
```

```python
getTitlesandImgUrls('https://movie.naver.com/movie/search/result.naver?section=movie&query=%EC%9D%B8%EC%96%B4%EA%B3%B5%EC%A3%BC&page=1')
```

```python
def getAllTitlesandImgUrlsByName(movieName):
    titlesAndImgUrls={}
    page=1
    while True:
        url = 'https://movie.naver.com/movie/search/result.naver?section=movie&query='+urllib.parse.quote(movieName)+'&page='+str(page)
        temp = getTitlesandImgUrls(url)
        beforesize = len(titlesAndImgUrls)
        titlesAndImgUrls.update(temp)
        aftersize = len(titlesAndImgUrls)
        if beforesize == aftersize:
            break
        page += 1
    return titlesAndImgUrls
```

```python
getAllTitlesandImgUrlsByName('스타워즈')
```

```python
def downloadImg(filename, imgUrl):
    img = requests.get(imgUrl, verify=False)
    img.raise_for_status()
    with open(filename, 'wb') as f:
        f.write(img.content)
    print(f"{filename} 다운로드 완료 {len(img.content)/1000}kB")
```

```python
def removeWhiteSpaceAndSlash(s):
    temp = s.replace('/','')
    temp = temp.replace(':','')
    temp = temp.replace(' ','')
    return temp
```

```python
import os
```

```python
def downloadMoviePoster(movieName):
    titlesAndUrls = getAllTitlesandImgUrlsByName(movieName)
    os.makedirs(movieName, exist_ok=True)
    for title in titlesAndUrls:
        if 'poster_default.gif' not in titlesAndUrls[title]:
            filename = './'+movieName+'/'+removeWhiteSpaceAndSlash(title)+'.jpg'
            url = titlesAndUrls[title]
            downloadImg(filename,url)
```

```python
downloadMoviePoster('스타워즈')
```

```python
downloadMoviePoster('인어공주')
```
