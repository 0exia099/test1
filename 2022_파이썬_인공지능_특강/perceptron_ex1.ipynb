```python


#학습 알고리즘
def train_weights(X, y, l_rate, n_epoch):
    global weights
    global bias
    for epoch in range(n_epoch):
        sum_error = 0.0
        for row, target in zip(X,y):
            actual = calculate(row)
            error = target - actual
            bias = bias + l_rate*error
            sum_error += error**2
            for i in range(2):
                weights[i] = weights[i] + l_rate*error*row[i]
            print(weights, bias)
        print('에포크 번호 = %d, 학습률 = %.3f, 오류 = %.3f' % (epoch, l_rate, sum_error))
    return weights
```

```python
#학습 데이터 셋
X = [[0,0],[0,1],[1,0],[1,1]]
y = [0,0,0,1]

#가중치와 바이어스 초기값
weights = [0.0,0.0]
bias = 0.0

l_rate = 0.1
n_epoch = 5
#테스트
weights = train_weights(X,y,l_rate, n_epoch)

print("\n\n학습 완료")
print(weights, bias)
```
