### Aaron Abrams와 함께 출연한 배우들의 리스트

```python
aaron = casts[casts.name == 'Aaron Abrams']
```

```python
aaron
```

```python
aaron.shape
```

```python
len(aaron)
```

```python
aaron.merge(casts) #aaron과 casts는 항목들이 같아 aaron과 같은 결과가 나온다
```

```python
costar = aaron.merge(casts, on = ['title', 'year']) #aaron과 제목과 연도가 같은 것을 찾음
#aaron이 출연한 영화의 제목과 연도가 같으면 같은 영화에 출연했다는 것
```

```python
costar #나머지 항목들이 다 두개가 된다. 그리고 aaron까지 같이 나옴
```

```python
costar = costar[costar.name_x != costar.name_y] #aaron이 두번나오는거 뺌
```

```python
costar
```

```python
costar[['title', 'year', 'name_y']] #같이 출연한 배우들과 영화 정보(한 영화에 배역을 여러개 맡아 두번 나오는 사람존재)
```

```python
costar = costar[['title', 'year', 'name_y']]
```
