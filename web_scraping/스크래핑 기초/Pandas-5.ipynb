### 2000년대에 개봉한 영화..(2000~2009)

```python
title2000 = titles[(titles.year>=2000) & (titles.year<=2009)]
```

```python
title2000.head()
```

```python
title2000
```

```python
title2000.sort_values(by = 'year') #같은 연도끼리는 아무렇게나 정렬되어 있음
```

```python
title2000.sort_values(['year', 'title']) #연도 정렬 후 그 다음 제목으로 정렬(대문자 후 소문자)
```

```python
title2000.sort_values(['year', 'title'], ascending = [False, True]) #연도는 내림차순, 제목은 오름차순으로 정렬
```
