```python
import pandas as pd
```

```python
data = {
    '이름' : ['홍길동', '임꺽정', '이순신'],
    '중간' : [40,60,100],
    '기말' : [60,50,95],
    '출석' : [8.5, 7.8,9.9]
}
```

```python
df = pd.DataFrame(data)
```

```python
df['주소'] = '경산'
```

```python
df['생일'] = ['1999-01-02', '2010-09-17', '2015-01-18']
```

```python
df
```

```python
newLines = [
    {'이름' : '전우치', '중간' : 70, '기말' : 80,'출석':9.1,'주소' : '대구', '생일' : '2012-03-05'},
    {'이름' : '이성계', '중간' : 44, '기말' : 89,'출석':7.4,'주소' : '대구', '생일' : '2008-03-19'}
]
```

```python
newLines
```

```python
df.append(newLines) #인덱스 번호가 맞지 않는다.
```

```python
df.append(newLines, ignore_index = True) #ignore_index = True로 해결
```

```python
df2 = df.append(newLines, ignore_index = True)
```
