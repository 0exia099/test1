### 다운로드 함수

```python
def download(url, filename, data): #데이터 받아오는 함수
    with open(filename, 'wb') as f:
        r = requests.post(url, data = data)
        if r.status_code != 200:
            print('다운로드 실패')
            return
        f.write(r.content)
        print('다운로드 성공 : ' + filename)
```

```python
def getFormData(startDay, endDay): #Form Data 설정 함수
    data = {
        'fileType': 'csv',
        'pgmNo': '70',
        'menuNo': '432',
        'serviceSe': 'F00101',
        'stdrMg': '99999',
        'startDt': startDay,#시작날짜,끝날짜 변수로 설정
        'endDt': endDay,
        'taElement': 'MIN',
        'taElement': 'AVG',
        'taElement': 'MAX',
        'stnGroupSns': '',
        'selectType': '1',
        'mddlClssCd': 'SFC01',
        'dataFormCd': 'F00501',
        'dataTypeCd': 'standard',
        'startDay': startDay,
        'startYear': '2022',
        'endDay': endDay,
        'endYear': '2022',
        'startMonth': '01',
        'endMonth': '12',
        'sesnCd': '0',
        'txtStnNm': '대구', #위치도 바꾸고싶으면 변수 하나 더
        'stnId': '143',
        'areaId': '',
        'gFontSize':'' 
    }
    return data
```

```python
def downloadTempCSV(startDay, endDay):#CSV받아오는 함수
    startDay = str(startDay) #startDay와 endDay를 문자열로 변경
    endDay = str(endDay)
    url = 'https://data.kma.go.kr/stcs/grnd/downloadGrndTaList.do' #기온정보 받는 곳 주소
    data = getFormData(startDay, endDay) #만든 getFormData함수로 data지정
    filename = 'temp_' + startDay + '_' + endDay + '.csv'
    download(url, filename, data) #만든 download함수로 csv파일에 저장할 정보 받아온다
```
